## Listas

### Gerando listas aleatórias

```python
l1 = [random.randint(10000, 99999) for i in range(100)]
l2 = [random.randint(10000, 99999) for i in range(100)]
l3 = l1.copy(); random.shuffle(l3)

lists = [l1, l2, l3]
```

```python
for i in range(3):
    print(f'L{i+1} = {lists[i]}')
```

### Terceiro maior elemento

```python
lists_copies = [lista.copy() for lista in lists]

for i in range(3):
    lists_copies[i].sort()
    print(f'Resultado L{i+1} = {third_biggest_element(lists[i])} | Real = {lists_copies[i][-3]}')
```

### Partição

```python
l12, l22 = l1.copy(), l2.copy()
```

```python
partition_lists(l12, l22)
```

```python
print(f'L1_2 = {l12}')
print(f'L2_2 = {l22}')
print(f'max(L1_2) = {max(l12)} | min(L2_2) = {min(l22)}')
print(min(l22) > max(l12))
```

### Permutações

```python
print(f'L1 e L2 são permutaçãoes? {are_lists_permutations(l1, l2)}')
print(f'L1 e L3 são permutaçãoes? {are_lists_permutations(l1, l3)}')
```
